```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv('/content/Titanic-Dataset.csv')
print(titanic.head())
print(titanic.shape)
```

```python
sns.barplot(data=titanic, x='Sex', y='Age')
plt.title('Barplot of Sex vs Age')
plt.show()
```

```python
alive_count = titanic[titanic['Survived'] == 1].shape[0]
print(f"Number of passengers alive: {alive_count}")

```

```python
# 4. How many male and female survived w.r.t to passenger class?
survived_gender_class = titanic[titanic['Survived'] == 1].groupby(['Pclass', 'Sex']).size()
print("Number of male and female survivors w.r.t passenger class:")
print(survived_gender_class)

```

```python
# 5. Make a distribution plot of fare making the color red of the bars
sns.histplot(titanic['Fare'], kde=False, color='red')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

```

```python
# 6. Make a jointplot of fare and age. Write down which age group is the highest in the graph?
sns.jointplot(data=titanic, x='Fare', y='Age', kind='scatter')
plt.title('Jointplot of Fare and Age', loc='center', pad=80)
plt.show()
```

```python
# 7. Make a boxplot of class and age
sns.boxplot(data=titanic, x='Pclass', y='Age')
plt.title('Boxplot of Class and Age')
plt.show()

```

```python
# Select only numeric columns before calculating correlation
numerical_titanic = titanic.select_dtypes(include=['number'])
correlation_matrix = numerical_titanic.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Data points
data = np.array([[7, 8], [3, 5], [6, 4], [6, 0], [1, 5], [8, 4]])

# Calculate linkage matrix using average linkage
Z = linkage(data, 'complete')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=list('ABCDEF'))
plt.title('Hierarchical Clustering (Average Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
