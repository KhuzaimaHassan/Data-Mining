```python
import random
import string

def generate_password():
    # Define the character pools
    upper_case = string.ascii_uppercase
    lower_case = string.ascii_lowercase
    digits = string.digits
    special_chars = "$_."

    # Ensure password includes at least one character from each pool
    password = [
        random.choice(upper_case),
        random.choice(lower_case),
        random.choice(digits),
        random.choice(special_chars)
    ]

    # Determine the total length of the password
    total_length = random.randint(8, 12)

    # Fill the remaining password length with random characters from all pools
    all_chars = upper_case + lower_case + digits + special_chars
    password += random.choices(all_chars, k=total_length - len(password))

    # Shuffle the password to avoid predictable patterns
    random.shuffle(password)

    # Join the characters into a single string
    return ''.join(password)

# Generate and print the random password
print("Generated Password:", generate_password())
```

```python
import pandas as pd
df_sal=pd.read_csv('/content/lab2_3_Salaries.csv')
```

```python
rows, columns = df_sal.shape
print(f"Rows: {rows}, Columns: {columns}")
```

```python
average_basepay = df_sal['BasePay'].mean()
print(f"Average BasePay: {average_basepay}")
```

```python
highest_overtimepay = df_sal['OvertimePay'].max()
print(f"Highest OvertimePay: {highest_overtimepay}")
```

```python
joseph_driscoll_job = df_sal[df_sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle'].values[0]
print(f"Job Title of JOSEPH DRISCOLL: {joseph_driscoll_job}")
```

```python
joseph_driscoll_total_pay = df_sal[df_sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits'].values[0]
print(f"JOSEPH DRISCOLL Total Pay (including benefits): {joseph_driscoll_total_pay}")
```

```python
highest_paid_person = df_sal[df_sal['TotalPayBenefits'] == df_sal['TotalPayBenefits'].max()]['EmployeeName'].values[0]
print(f"Highest Paid Person: {highest_paid_person}")
```

```python
lowest_paid_person = df_sal[df_sal['TotalPayBenefits'] == df_sal['TotalPayBenefits'].min()]['EmployeeName'].values[0]
lowest_paid_amount = df_sal['TotalPayBenefits'].min()
print(f"Lowest Paid Person: {lowest_paid_person}, Amount: {lowest_paid_amount}")
```

```python
average_basepay_per_year = df_sal.groupby('Year')['BasePay'].mean()
print("Average BasePay per Year:")
print(average_basepay_per_year)
```

```python
top_5_jobs = df_sal['JobTitle'].value_counts().head(5)
print("Top 5 Most Common Jobs:")
print(top_5_jobs)
```

```python
job_titles_2013 = df_sal[df_sal['Year'] == 2013]['JobTitle'].value_counts()
unique_jobs_2013 = (job_titles_2013 == 1).sum()
print(f"Job Titles Represented by Only One Person in 2013: {unique_jobs_2013}")
```
